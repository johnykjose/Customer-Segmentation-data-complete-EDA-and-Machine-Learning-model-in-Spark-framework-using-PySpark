## Data Preprocessing

```python
#Before string Indexing
cat_cols = ['Month','VisitorType']
pd.DataFrame(data_df.take(5), columns=data_df.columns)
```

```python
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler

stages = []
for categoricalCol in cat_cols:
    # Category Indexing with StringIndexer
    stringIndexer = StringIndexer(inputCol=categoricalCol, outputCol=categoricalCol + "Index")
    # Use OneHotEncoder to convert categorical variables into binary SparseVectors
    encoder = OneHotEncoder(inputCols=[stringIndexer.getOutputCol()], outputCols=[categoricalCol + "classVec"])
    # Add stages.  These are not run here, but will run all at once later on.
    stages += [stringIndexer, encoder]
```

```python
#label_stringIdx = StringIndexer(inputCol = 'Revenue', outputCol = 'Revenue')
#stages += [label_stringIdx]
```

```python
assemblerInputs = [c + "classVec" for c in cat_cols] + num_cols + bool_cols
assembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
stages += [assembler]

#standardization
standardScaler = StandardScaler(inputCol="features", outputCol="features_scaled")
stages += [standardScaler]
```

```python
#pipeline
from pyspark.ml import Pipeline
pipeline = Pipeline(stages = stages)
pipelineModel = pipeline.fit(data_df)
df = pipelineModel.transform(data_df)
pd.DataFrame(df.take(5), columns=df.columns)
```

```python
selectedCols = ['Revenue', 'features_scaled'] + data_df.columns
df = df.select(selectedCols)
df.printSchema()
```

```python
# Split the data into train and test sets
train, test = df.randomSplit([.75,.25], seed=rnd_seed)
print("Training Dataset Count: " + str(train.count()))
print("Test Dataset Count: " + str(test.count()))
```

```python
# LOAD PYSPARK LIBRARIES
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from sklearn.metrics import roc_curve,auc
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import BinaryClassificationMetrics as metric

def model_evaluation(model_predictions,model):
    print('Accuracy: ', MulticlassClassificationEvaluator(labelCol='label',metricName='accuracy').evaluate(model_predictions))
    print('Precision: ',MulticlassClassificationEvaluator(labelCol='label',metricName='weightedPrecision').evaluate(model_predictions))
    print('F1-score: ',MulticlassClassificationEvaluator(labelCol='label',metricName='truePositiveRateByLabel').evaluate(model_predictions))
    if model == 'svc':
        results = model_predictions.select(['rawPrediction', 'label'])
    else:
        results = model_predictions.select(['probability', 'label'])
    ## prepare score-label set
    results_collect = results.collect()
    results_list = [(float(i[0][0]), 1.0-float(i[1])) for i in results_collect]
    scoreAndLabels = sc.parallelize(results_list)
    metrics = metric(scoreAndLabels)
    print("The ROC score is : ", metrics.areaUnderROC)
    from sklearn.metrics import roc_curve, auc
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    y_test = [i[1] for i in results_list]
    y_score = [i[0] for i in results_list]
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    %matplotlib inline
    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Area under the ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
    #Confusion matrix
    from sklearn.metrics import confusion_matrix
    y_true = model_predictions.select("label")
    y_true = y_true.toPandas()
    y_pred = model_predictions.select("prediction")
    y_pred = y_pred.toPandas()
    sns.heatmap(confusion_matrix(y_pred,y_true),annot=True,cbar=False,fmt = '.1f',cmap = 'Blues')
    title = 'Test Confusion Matrix'
    plt.title(title,fontsize=18)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    #print(pd.DataFrame(model_predictions.take(5), columns=model_predictions.columns))
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit

# DEFINE ALGORITHM 
def build_LR(train,test):
    lr = LogisticRegression(labelCol="label", featuresCol="features_scaled")

    # DEFINE GRID PARAMETERS
    paramGrid = ParamGridBuilder().addGrid(lr.regParam, (0.01, 0.1))\
                                  .addGrid(lr.maxIter, (5, 10))\
                                  .addGrid(lr.tol, (1e-4, 1e-5))\
                                  .addGrid(lr.elasticNetParam, (0.25,0.75))\
                                  .build()

    # DEFINE CROSS VALIDATION WITH PARAMETERS
    tvs = TrainValidationSplit( estimator=lr
                               ,estimatorParamMaps=paramGrid
                               ,evaluator=BinaryClassificationEvaluator(labelCol='label')
                               ,trainRatio=0.8)
    model = tvs.fit(train)
    model_predictions= model.transform(test)
    return model_predictions


def build_RF(train,test):
    rf = RandomForestClassifier(labelCol='label',featuresCol="features_scaled")
    paramGrid = ParamGridBuilder()\
                                    .addGrid(rf.maxDepth, [5, 10, 20]) \
                                    .addGrid(rf.maxBins, [20, 32, 50]) \
                                    .addGrid(rf.numTrees, [20, 40, 60 ]) \
                                    .addGrid(rf.impurity, ["gini", "entropy"]) \
                                    .addGrid(rf.minInstancesPerNode, [1, 5, 10]) \
                                    .build()


    tvs = TrainValidationSplit( estimator=rf
                               ,estimatorParamMaps=paramGrid
                               ,evaluator=MulticlassClassificationEvaluator(labelCol='label')
                               ,trainRatio=0.8)
    model = tvs.fit(train)
    model_predictions= model.transform(test)
    return model_predictions

def build_SVC(train,test):
    from pyspark.ml.classification import LinearSVC
    rf = LinearSVC(labelCol='label',featuresCol="features_scaled")
    paramGrid = ParamGridBuilder()\
                                    .addGrid(rf.regParam, (0.01, 0.1)) \
                                    .addGrid(rf.tol, (1e-4, 1e-5)) \
                                    .addGrid(rf.maxIter, (5, 10)) \
                                    .build()


    tvs = TrainValidationSplit( estimator=rf
                               ,estimatorParamMaps=paramGrid
                               ,evaluator=MulticlassClassificationEvaluator(labelCol='label')
                               ,trainRatio=0.8)
    model = tvs.fit(train)
    model_predictions= model.transform(test)
    return model_predictions
```

```python
### Logistic Regression
model_evaluation(build_LR(train,test),'lr')
```

```python
### Randome Forest
model_evaluation(build_RF(train,test),'rf')
```

```python
### Linear SVC
model_evaluation(build_SVC(train,test),'svc')
```
